```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv(r'C:\Users\M.Eren\Desktop\PRSA.csv')
```

```python
data.head()
```

```python
data.info()
data.describe()
```

```python
data.isnull().sum()
```

```python
data['PM2.5'] = data['PM2.5'].fillna(data['PM2.5'].median())
data['PM10'] = data['PM10'].fillna(data['PM10'].median())
data['SO2'] = data['SO2'].fillna(data['SO2'].median())
data['NO2'] = data['NO2'].fillna(data['NO2'].median())
data['CO'] = data['CO'].fillna(data['CO'].median())
data['O3'] = data['O3'].fillna(data['O3'].median())
data['TEMP'] = data['TEMP'].fillna(data['TEMP'].median())
data['PRES'] = data['PRES'].fillna(data['PRES'].median())
data['DEWP'] = data['DEWP'].fillna(data['DEWP'].median())
data['RAIN'] = data['RAIN'].fillna(data['RAIN'].median())
data['WSPM'] = data['WSPM'].fillna(data['WSPM'].median())
data['wd'] = data['wd'].fillna(data['wd'].mode()[0])
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data['PM2.5'])
plt.show()
```

```python
q1 = data['PM2.5'].quantile(0.25)
q3 = data['PM2.5'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
data = data[(data['PM2.5'] >= lower_bound) & (data['PM2.5'] <= upper_bound)]
```

```python
features = data[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']]  # Örnek özellikler
target = data['PM2.5']  
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='month', y='PM2.5', data=data)
plt.title('Aylara Göre PM2.5 Seviyesi')
plt.xlabel('Ay')
plt.ylabel('PM2.5 Seviyesi')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TEMP', y='PM2.5', data=data)
plt.title('Sıcaklık ve PM2.5 Seviyesi Arasındaki İlişki')
plt.xlabel('Sıcaklık (°C)')
plt.ylabel('PM2.5 Seviyesi')
plt.show()
```
